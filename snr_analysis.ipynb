```python
import pandas as pd

lass_df = pd.read_csv("lass_synthetic_validation.csv")
onepeace_df = pd.read_csv("results/one-peace_validation_results.csv")

lass_df['source_file'] = lass_df['source'].apply(lambda x: f"lass_validation/{x}.wav")
lass_df['noise_file'] = lass_df[' noise'].apply(lambda x: f"lass_validation/{x}.wav")

merged_df = pd.merge(
    lass_df,
    onepeace_df,
    left_on=['source_file', 'noise_file'],
    right_on=['source_path', 'noise_path'],
    how='inner'
)

merged_df = merged_df.drop(columns=['source_file', 'noise_file', 'caption'])

pd.set_option("display.max_colwidth", None)

# Display the first few rows of the merged DataFrame
display(merged_df.head())
```
