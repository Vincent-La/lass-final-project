```python
import pandas as pd
import matplotlib.pyplot as plt

baseline_path = 'results/baseline_validation_results.csv'
baseline_df = pd.read_csv(baseline_path)
one_peace_path = 'results/one-peace_validation_results.csv'
one_peace_df = pd.read_csv(one_peace_path)

pd.set_option("display.max_colwidth", None)
```

```python
display(one_peace_df.head())
```

```python
# Sort by target_similarity and show the lowest/highest 5 rows
lowest_target_similarity = one_peace_df.nsmallest(5, 'target_similarity')
highest_target_similarity = one_peace_df.nlargest(5, 'target_similarity')

print("Lowest 5 rows by target_similarity:")
display(lowest_target_similarity)

print("Highest 5 rows by target_similarity:")
display(highest_target_similarity)
```

```python
# Sort by output_similarity and show the lowest/highest 5 rows
lowest_output_similarity = one_peace_df.nsmallest(5, 'output_similarity')
highest_output_similarity = one_peace_df.nlargest(5, 'output_similarity')

print("Lowest 5 rows by output_similarity:")
display(lowest_output_similarity)

print("Highest 5 rows by output_similarity:")
display(highest_output_similarity)
```

```python
# Sort by delta_similarity and show the lowest/highest 5 rows
lowest_delta_similarity = one_peace_df.nsmallest(5, 'delta_similarity')
highest_delta_similarity = one_peace_df.nlargest(5, 'delta_similarity')

print("Lowest 5 rows by delta_similarity:")
display(lowest_delta_similarity)

print("Highest 5 rows by delta_similarity:")
display(highest_delta_similarity)
```

```python
# Compute delta_similarity + sdri and delta_similarity - sdri, and sort
# Add these as new columns to the dataframe
sim_weight = 100
one_peace_df['delta_plus_sdri'] = one_peace_df['delta_similarity'] * sim_weight + one_peace_df['sdri']
one_peace_df['delta_minus_sdri'] = one_peace_df['delta_similarity'] * sim_weight - one_peace_df['sdri']

# Sort by delta_similarity + sdri
lowest_delta_plus_sdri = one_peace_df.nsmallest(5, 'delta_plus_sdri')
highest_delta_plus_sdri = one_peace_df.nlargest(5, 'delta_plus_sdri')

print("Lowest 5 rows by delta_similarity + sdri:")
display(lowest_delta_plus_sdri)

print("Highest 5 rows by delta_similarity + sdri:")
display(highest_delta_plus_sdri)
```

```python
# Sort by delta_similarity - sdri
lowest_delta_minus_sdri = one_peace_df.nsmallest(5, 'delta_minus_sdri')
highest_delta_minus_sdri = one_peace_df.nlargest(5, 'delta_minus_sdri')

print("Lowest 5 rows by delta_similarity - sdri:")
display(lowest_delta_minus_sdri)

print("Highest 5 rows by delta_similarity - sdri:")
display(highest_delta_minus_sdri)
```

```python
# Plot correlation between delta_similarity and delta_sdr
plt.figure(figsize=(8, 6))
plt.scatter(one_peace_df['delta_similarity'], one_peace_df['sdri'], alpha=0.5, s=10)
plt.title('ONE-PEACE DeltaSimilarity vs SDRi')
plt.xlabel('DeltaSimilarity')
plt.ylabel('SDRi')
plt.savefig('plots/similarity_vs_sdr/onepeace_deltasim_sdri.png', dpi=600)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(baseline_df['delta_similarity'], baseline_df['sdri'], alpha=0.5, s=10)
plt.title('Baseline DeltaSimilarity vs SDRi')
plt.xlabel('DeltaSimilarity')
plt.ylabel('SDRi')
plt.savefig('plots/similarity_vs_sdr/baseline_deltasim_sdri.png', dpi=600)
plt.show()
```

```python
plot_params = [['delta_similarity', 'sdr', 'DeltaSimilarity', 'SDR', 'ONE-PEACE', 'onepeace_deltasim_sdr'],
               ['delta_similarity', 'sisdr', 'DeltaSimilarity', 'SI-SDR', 'ONE-PEACE', 'onepeace_deltasim_sisdr'],
               ['delta_similarity', 'sdr', 'OutputSimilarity', 'SDR', 'ONE-PEACE', 'onepeace_outputsim_sdr'],
               ['delta_similarity', 'sisdr', 'OutputSimilarity', 'SI-SDR', 'ONE-PEACE', 'onepeace_outputsim_sisdr'],
               ['delta_similarity', 'sdr', 'DeltaSimilarity', 'SDR', 'Baseline', 'baseline_deltasim_sdr'],
               ['delta_similarity', 'sisdr', 'DeltaSimilarity', 'SI-SDR', 'Baseline', 'baseline_deltasim_sisdr'],
               ['delta_similarity', 'sdr', 'OutputSimilarity', 'SDR', 'Baseline', 'baseline_outputsim_sdr'],
               ['delta_similarity', 'sisdr', 'OutputSimilarity', 'SI-SDR', 'Baseline', 'baseline_outputsim_sisdr']]

for col1, col2, xlabel, ylabel, embedding, filename in plot_params:
    plt.figure(figsize=(8, 6))
    if embedding == 'ONE-PEACE':
        plt.scatter(one_peace_df[col1], one_peace_df[col2], alpha=0.5, s=10)
    else:
        plt.scatter(baseline_df[col1], baseline_df[col2], alpha=0.5, s=10)
    plt.title(f'{embedding} {xlabel} vs {ylabel}')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.savefig(f'plots/similarity_vs_sdr/{filename}.png', dpi=600)
```
