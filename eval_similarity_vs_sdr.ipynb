```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a Pandas DataFrame
file_path = 'results/one-peace_validation_results.csv'
df = pd.read_csv(file_path)
```

```python
display(df.head())
```

```python
# Sort by target_similarity and show the lowest/highest 5 rows
lowest_target_similarity = df.nsmallest(5, 'target_similarity')
highest_target_similarity = df.nlargest(5, 'target_similarity')

print("Lowest 5 rows by target_similarity:")
display(lowest_target_similarity)

print("Highest 5 rows by target_similarity:")
display(highest_target_similarity)
```

```python
# Sort by output_similarity and show the lowest/highest 5 rows
lowest_output_similarity = df.nsmallest(5, 'output_similarity')
highest_output_similarity = df.nlargest(5, 'output_similarity')

print("Lowest 5 rows by output_similarity:")
display(lowest_output_similarity)

print("Highest 5 rows by output_similarity:")
display(highest_output_similarity)
```

```python
# Sort by delta_similarity and show the lowest/highest 5 rows
lowest_delta_similarity = df.nsmallest(5, 'delta_similarity')
highest_delta_similarity = df.nlargest(5, 'delta_similarity')

print("Lowest 5 rows by delta_similarity:")
display(lowest_delta_similarity)

print("Highest 5 rows by delta_similarity:")
display(highest_delta_similarity)
```

```python
# Compute delta_similarity + sdri and delta_similarity - sdri, and sort
# Add these as new columns to the dataframe
sim_weight = 100
df['delta_plus_sdri'] = df['delta_similarity'] * sim_weight + df['sdri']
df['delta_minus_sdri'] = df['delta_similarity'] * sim_weight - df['sdri']

# Sort by delta_similarity + sdri
lowest_delta_plus_sdri = df.nsmallest(5, 'delta_plus_sdri')
highest_delta_plus_sdri = df.nlargest(5, 'delta_plus_sdri')

print("Lowest 5 rows by delta_similarity + sdri:")
display(lowest_delta_plus_sdri)

print("Highest 5 rows by delta_similarity + sdri:")
display(highest_delta_plus_sdri)
```

```python
# Sort by delta_similarity - sdri
lowest_delta_minus_sdri = df.nsmallest(5, 'delta_minus_sdri')
highest_delta_minus_sdri = df.nlargest(5, 'delta_minus_sdri')

print("Lowest 5 rows by delta_similarity - sdri:")
display(lowest_delta_minus_sdri)

print("Highest 5 rows by delta_similarity - sdri:")
display(highest_delta_minus_sdri)
```

```python
# Plot correlation between delta_similarity and delta_sdr
plt.figure(figsize=(8, 6))
plt.scatter(df['delta_similarity'], df['sdri'], alpha=0.5, s=10)
plt.title('Correlation between delta_similarity and sdri')
plt.xlabel('delta_similarity')
plt.ylabel('sdri')
plt.grid(True)
plt.show()
```
